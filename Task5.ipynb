### Problem 1: Social Media – Post Engagement Analyzer

```python
def analyze_post_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {post_status}")

likes_data_viral = [1] * 100000 #1M views and likes
analyze_post_engagement(likes_data_viral)

print();
likes_data_normal = [1] * 500 #Normal views
analyze_post_engagement(likes_data_normal)
```

### Problem 2: Healthcare – Medicine Stock Alert System

```python
def check_medicine_stock(stock_level):
    if stock_level < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    print(f"Medicine Stock: {stock_level}")
    print(f"Status: {status}")

check_medicine_stock(6) # Low Stock Scenerio
print();
check_medicine_stock(15) # Sufficient Stock Scenario
print();
check_medicine_stock(10) # Borderline Stock Scenario
```

## Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def analyze_rainfall_adequacy(rainfall_data, required_level):
    total_rainfall = sum(rainfall_data)
    num_months = len(rainfall_data)

    if num_months > 0:
        average_rainfall = total_rainfall / num_months
    else:
        average_rainfall = 0

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {average_rainfall:.0f}")
    print(f"Rainfall Status: {status}")

seasonal_rainfall_adequate = [60, 80, 70, 75, 65]
required_level_adequate = 70
print("Scenario 1: Adequate Rainfall")
analyze_rainfall_adequacy(seasonal_rainfall_adequate, required_level_adequate)

print();

seasonal_rainfall_inadequate = [40, 50, 30, 45, 55]
required_level_inadequate = 60
print("Scenario 2: Inadequate Rainfall")
analyze_rainfall_adequacy(seasonal_rainfall_inadequate, required_level_inadequate)
```

## Problem 4: Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    unique_usernames = set()
    has_duplicates = False

    for username in usernames:
        if username in unique_usernames:
            has_duplicates = True
            break
        unique_usernames.add(username)

    if has_duplicates:
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

print("Scenario 1: With Duplicates")
user_list_with_duplicates = ["user1", "user2", "user1", "user3"]
detect_duplicate_accounts(user_list_with_duplicates)

print();

print("Scenario 2: Without Duplicates")
user_list_unique = ["userA", "userB", "userC", "userD"]
detect_duplicate_accounts(user_list_unique)
```

## Problem 5: Healthcare – Appointment Eligibility Checker

```python
def check_appointment_eligibility(age):
    if age >= 18:
        eligibility_status = "Eligible"
    else:
        eligibility_status = "Not Eligible"
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {eligibility_status}")

print("Scenario 1: Eligible Patient")
check_appointment_eligibility(21)
print();
print("Scenario 2: Not Eligible Patient")
check_appointment_eligibility(16)
print();
print("Scenario 3: Borderline Eligible Patient")
check_appointment_eligibility(18)
```

## Problem 6: Agriculture – Premium Crop Price Filter

```python
def filter_premium_crops(crop_prices):
    premium_threshold = 2000
    premium_crops = []

    for price in crop_prices:
        if price > premium_threshold:
            premium_crops.append(price)

    print(f"Premium Crops: {premium_crops}")

crop_data = [1500, 2500, 1800, 3200, 1950]
filter_premium_crops(crop_data)
```

## Problem 7: System Monitoring – Application Health Checker

```python
def check_application_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif 1 <= error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

print("Scenario 1: Healthy System")
check_application_health(0)
print();
print("Scenario 2: Minor Issues")
check_application_health(3)
```

## Problem 8: Banking – Daily Transaction Limit Checker

```python
def check_transaction_limit(transaction_amount):
    daily_limit = 50000
    if transaction_amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    print(f"Transaction Amount: {transaction_amount}")
    print(f"Transaction Status: {status}")

print("Scenario 1: Approved Transaction")
check_transaction_limit(45000)
print();
print("Scenario 2: Rejected Transaction")
check_transaction_limit(60000)
print();
print("Scenario 3: Borderline Approved Transaction")
check_transaction_limit(50000)
```

## Problem 9: E-Learning – Student Attendance Eligibility System

```python
def check_exam_eligibility(attended_classes, total_classes):
    if total_classes == 0:
        attendance_percentage = 0.0
    else:
        attendance_percentage = (attended_classes / total_classes) * 100

    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {eligibility}")

print("Scenario 1: Eligible Student")
check_exam_eligibility(75, 100)
print();
print("Scenario 2: Not Eligible Student")
check_exam_eligibility(70, 100)
print();
print("Scenario 3: Borderline Eligible Student")
check_exam_eligibility(30, 40)
```

## Problem 10: Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units_consumed):
    total_bill = 0
    usage_status = ""
    if units_consumed <= 100:
        total_bill = units_consumed * 3
    elif units_consumed <= 200:
        total_bill = (100 * 3) + ((units_consumed - 100) * 5)
    else:
        total_bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)
    if total_bill < 500:
        usage_status = "Low Usage"
    elif 500 <= total_bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    print(f"Units Consumed: {units_consumed}")
    print(f"Total Bill: ₹{total_bill:.2f}")
    print(f"Usage Status: {usage_status}")
print("Scenario 1: Low Usage (50 units)")
calculate_electricity_bill(50)
print();
print("Scenario 2: Moderate Usage (150 units)")
calculate_electricity_bill(150)
print();
print("Scenario 3: High Usage (300 units)")
calculate_electricity_bill(300)
print();
print("Scenario 4: Borderline Low Usage (166 units -> 300 + 330 = 630)")
calculate_electricity_bill(166)
print();
print("Scenario 5: Borderline High Usage (200 units -> 300 + 500 = 800)")
calculate_electricity_bill(200)
```

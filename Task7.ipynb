### Problem 1: Smart Light Controller

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def display_status(self):
        status = "ON" if self.is_on else "OFF"
        print(f"{self.name} is {status}")

light1 = SmartLight("Bedroom Light")
print(f"Light Name: {light1.name}")
print("Action: ON")
light1.turn_on()
light1.display_status()
```

### Problem 2: Employee ID Card System

```python
class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card_details(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

employee1 = EmployeeIDCard("Rahul", "EMP102", "IT")
print(f"Employee Name: {employee1.name}")
print(f"Employee ID: {employee1.emp_id}")
print(f"Department: {employee1.department}")
employee1.display_card_details()
```

### Problem 3: Mobile Contact Record

```python
class ContactRecord:
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def display_contact_info(self):
        print(f"Contact Saved\nName: {self.name} Phone: {self.phone_number}")

contact1 = ContactRecord("Anita", "9876543210")
print(f"Contact Name: {contact1.name}")
print(f"Phone Number: {contact1.phone_number}")

contact1.display_contact_info()
```

### Problem 4: Product Price Tag Generator

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

product1 = Product("Headphones", 2499)
print(f"Product Name: {product1.name}")
print(f"Price: {product1.price}")
product1.print_price_tag()
```

### Problem 5: Movie Rating Display System

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

movie1 = Movie("Inception", 4.8)
print(f"Movie Name: {movie1.name}")
print(f"Rating: {movie1.rating}")
movie1.display_movie_details()
```

### Problem 6: Delivery Address Manager

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

delivery1 = Delivery("Suman", "Hyderabad")
print(f"Customer Name: {delivery1.customer_name}")
print(f"Address: {delivery1.address}")
delivery1.print_delivery_details()
```

## 1. Smart Parking Lot Management System


```python
def manage_parking_lot(capacity, vehicle_logs):
    parked_vehicles = 0
    peak_parking_usage = 0
    print(f"Parking Capacity: {capacity}\n")
    for i, log in enumerate(vehicle_logs):
        if log == "IN":
            parked_vehicles += 1
            print(f"Log {i+1}: Vehicle IN. Current parked vehicles: {parked_vehicles}")
            if parked_vehicles > capacity:
                print(f"ALERT: Parking capacity exceeded! Current vehicles: {parked_vehicles}, Capacity: {capacity}")
        elif log == "OUT":
            if parked_vehicles > 0:
                parked_vehicles -= 1
                print(f"Log {i+1}: Vehicle OUT. Current parked vehicles: {parked_vehicles}")
            else:
                print(f"Log {i+1}: Vehicle OUT. No vehicles to exit, parked vehicles is already 0.")

        peak_parking_usage = max(peak_parking_usage, parked_vehicles)

    print(f"\nCurrently Parked Vehicles: {parked_vehicles}")

    if parked_vehicles == capacity:
        parking_status = "Full"
    elif parked_vehicles > capacity:
        parking_status = "Over Capacity"
    else:
        parking_status = "Available"

    print(f"Parking Status: {parking_status}")
    print(f"Peak Parking Usage: {peak_parking_usage}")

    return parked_vehicles, peak_parking_usage, parking_status
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

print("Test 1")
current, peak, status = manage_parking_lot(parking_capacity, vehicle_logs)

parking_capacity_small = 2
vehicle_logs_exceed = ["IN", "IN", "IN", "OUT", "IN"]

print("Testt 2")
current, peak, status = manage_parking_lot(parking_capacity_small, vehicle_logs_exceed)

parking_capacity_empty = 10
vehicle_logs_empty_exit = ["IN", "OUT", "OUT", "IN"]

print("Test 3")
current, peak, status = manage_parking_lot(parking_capacity_empty, vehicle_logs_empty_exit)
```

##2.Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance_km, weather_condition, traffic_level):
    base_time = distance_km * 3

    weather_delay = 0
    if weather_condition.lower() == 'rainy':
        weather_delay = 10
    elif weather_condition.lower() == 'snowy':
        weather_delay = 20
    elif weather_condition.lower() == 'stormy':
        weather_delay = 30

    traffic_delay = 0
    if traffic_level.lower() == 'medium':
        traffic_delay = 10
    elif traffic_level.lower() == 'high':
        traffic_delay = 20

    total_delivery_time = base_time + weather_delay + traffic_delay

    print(f"Distance: {distance_km} km")
    print(f"Weather: {weather_condition}")
    print(f"Traffic: {traffic_level}")
    print(f"Estimated Delivery Time: {int(total_delivery_time)} minutes\n")

    return int(total_delivery_time)

print("Test Case 1")
distance = 8
traffic = "High"
weather = "Rainy"
estimate_delivery_time(distance, weather, traffic)

print("est Case 2")
distance_ideal = 5
traffic_ideal = "Low"
weather_ideal = "Sunny"
estimate_delivery_time(distance_ideal, weather_ideal, traffic_ideal)

print("Test Case 3")
distance_bad = 15
traffic_bad = "Medium"
weather_bad = "Snowy"
estimate_delivery_time(distance_bad, weather_bad, traffic_bad)
```

## 3. Movie Theatre Seat Occupancy Analyzer

```python
def analyze_seat_occupancy(total_seats, booked_seats_list):
    num_booked_seats = len(booked_seats_list)
    occupancy_percentage = (num_booked_seats / total_seats) * 100

    show_status = "Available"
    if occupancy_percentage == 100:
        show_status = "Housefull"
    elif occupancy_percentage >= 80:
        show_status = "Almost Full"
    elif occupancy_percentage >= 50:
        show_status = "Moderately Occupied"

    suggestion = "No additional shows needed"
    if occupancy_percentage >= 90:
        suggestion = "Consider opening additional shows due to high demand"

    print(f"Total Seats: {total_seats}")
    print(f"Booked Seats: {num_booked_seats}")
    print(f"Occupancy: {occupancy_percentage:.2f}%")
    print(f"Show Status: {show_status}")
    print(f"Suggestion: {suggestion}\n")

    return occupancy_percentage, show_status, suggestion

print("--- Test Case 1: Sample Input (Almost Full) ---")
total_seats_1 = 200
booked_seats_1 = [1] * 150 # 150 booked seats
analyze_seat_occupancy(total_seats_1, booked_seats_1)

print("--- Test Case 2: Housefull Show ---")
total_seats_2 = 100
booked_seats_2 = [1] * 100 # 100 booked seats
analyze_seat_occupancy(total_seats_2, booked_seats_2)

print("--- Test Case 3: Low Occupancy ---")
total_seats_3 = 50
booked_seats_3 = [1] * 10 # 10 booked seats
analyze_seat_occupancy(total_seats_3, booked_seats_3)

print("--- Test Case 4: Moderately Occupied ---")
total_seats_4 = 100
booked_seats_4 = [1] * 65 # 65 booked seats
analyze_seat_occupancy(total_seats_4, booked_seats_4)
```

## 4. Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    if not cpu_readings:
        print("No CPU readings provided.")
        return "Unknown"

    total_cpu = sum(cpu_readings)
    average_cpu = total_cpu / len(cpu_readings)

    server_status = "Normal"
    if average_cpu >= 80:
        server_status = "Critical"
    elif average_cpu >= 50:
        server_status = "Warning"

    print(f"Average CPU Load: {average_cpu:.0f}%")
    print(f"Server Status: {server_status}\n")

    return server_status

print("--- Test Case 1: Sample Input (Warning) ---")
cpu_readings_1 = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings_1)

print("--- Test Case 2: Normal Load ---")
cpu_readings_2 = [20, 30, 40, 45]
classify_server_load(cpu_readings_2)

print("--- Test Case 3: Critical Load ---")
cpu_readings_3 = [80, 85, 90, 95]
classify_server_load(cpu_readings_3)

print("--- Test Case 4: Edge Case (Exactly 50%) ---")
cpu_readings_4 = [50, 49, 51]
classify_server_load(cpu_readings_4)

print("--- Test Case 5: Empty Readings ---")
cpu_readings_5 = []
classify_server_load(cpu_readings_5)
```

## 5. Smart Classroom Resource Usage Monitor

```python
def monitor_classroom_usage(resource_usage):
    overuse_thresholds = {
        "Projector": 5,
        "AC": 8,
        "Lights": 6
    }

    overused_resources = []
    energy_alert = "No"

    print("Resource Usage (hours):")
    for resource, usage_hours in resource_usage.items():
        print(f" {resource}: {usage_hours}")
        if resource in overuse_thresholds and usage_hours > overuse_thresholds[resource]:
            overused_resources.append(resource)
            energy_alert = "Yes"

    print(f"\nOverused Resources: {', '.join(overused_resources) if overused_resources else 'None'}")
    print(f"Energy Alert: {energy_alert}\n")

    return overused_resources, energy_alert


print("--- Test Case 1: Sample Input ---")
sample_usage_1 = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_classroom_usage(sample_usage_1)

print("--- Test Case 2: Normal Usage ---")
sample_usage_2 = {
    "Projector": 3,
    "AC": 7,
    "Lights": 5
}
monitor_classroom_usage(sample_usage_2)

print("--- Test Case 3: Multiple Overused Resources ---")
sample_usage_3 = {
    "Projector": 7,
    "AC": 8,
    "Lights": 7
}
monitor_classroom_usage(sample_usage_3)
```

## 6. Online Event Registration Capacity Controller

```python
def manage_event_registrations(event_capacity, registrations):
    confirmed_registrations = min(event_capacity, registrations)
    waitlisted_users = max(0, registrations - event_capacity)

    if confirmed_registrations < event_capacity:
        registration_status = "Open"
    elif confirmed_registrations == event_capacity and waitlisted_users == 0:
        registration_status = "Closed"
    else:
        registration_status = "Waitlist Full"

    print(f"Event Capacity: {event_capacity}")
    print(f"Total Registrations Attempted: {registrations}")
    print(f"Confirmed Registrations: {confirmed_registrations}")
    print(f"Waitlisted Users: {waitlisted_users}")
    print(f"Registration Status: {registration_status}\n")

    return confirmed_registrations, waitlisted_users, registration_status

print("--- Test Case 1: Sample Input (Overbooked) ---")
event_capacity_1 = 100
registrations_1 = 105
manage_event_registrations(event_capacity_1, registrations_1)

print("--- Test Case 2: Under Capacity ---")
event_capacity_2 = 50
registrations_2 = 30
manage_event_registrations(event_capacity_2, registrations_2)

print("--- Test Case 3: Exactly at Capacity ---")
event_capacity_3 = 75
registrations_3 = 75
manage_event_registrations(event_capacity_3, registrations_3)

print("--- Test Case 4: Significantly Over Capacity ---")
event_capacity_4 = 20
registrations_4 = 45
manage_event_registrations(event_capacity_4, registrations_4)
```

## 1.  Mobile Recharge Validation System



```python
def validate_recharge(amount, plan):
    if amount < 50:
        return False, "Recharge amount must be at least ₹50."
    valid_plans = {199, 299, 399, 599}
    if plan not in valid_plans:
        return False, f"Invalid plan. Valid plans are: {', '.join(map(str, valid_plans))}."
    return True, "Recharge successful!"

while True:
    recharge_amount = int(input("Enter recharge amount (in ₹): "))
    recharge_plan = int(input("Enter recharge plan: "))

    is_valid, message = validate_recharge(recharge_amount, recharge_plan)

    if is_valid:
        print(message)
        break
    else:
        print(f"Validation failed: {message}")
        print("Please try again.\n")
```

## 2. Inventory Reorder Alert System

```python
def check_inventory_reorder(products_stock):
    print("\n--- Inventory Reorder Status ---")
    for product, stock in products_stock.items():
        if stock < 15:
            print(f"Product: {product}, Stock: {stock} -> Reorder Alert!")
        else:
            print(f"Product: {product}, Stock: {stock} -> Stock OK")

products_inventory = {
    "Laptop": 10,
    "Mouse": 50,
    "Keyboard": 12,
    "Monitor": 25,
    "Webcam": 5
}

check_inventory_reorder(products_inventory)
```

## 3. Student Result Processing System


```python
def process_student_results(marks):
    if not marks:
        print("No marks provided.")
        return
    total_marks = sum(marks)
    average_marks = total_marks / len(marks)
    print(f"Student marks: {marks}")
    print(f"Average marks: {average_marks:.2f}")
    if average_marks >= 50:
        print("Result: Pass")
    else:
        print("Result: Fail")

student_marks_1 = [75, 80, 65, 90, 70]
print("Student 1 Results")
process_student_results(student_marks_1)
print()
student_marks_2 = [30, 45, 50, 40, 35]
print("Student 2 Results")
process_student_results(student_marks_2)
print()
student_marks_3 = [50, 50, 50, 50, 50]
print("Student 3 Results")
process_student_results(student_marks_3)
print()
student_marks_4 = []
print("Student 4 Results")
process_student_results(student_marks_4)
```

## 4. Cab Fare Estimator with Retry Option

```python
def calculate_cab_fare(distance_km, is_peak_hour):
    base_fare = 50
    fare_per_km = 12
    total_fare = base_fare + (distance_km * fare_per_km)
    if is_peak_hour:
        total_fare *= 1.25
    return total_fare

while True:
    distance_input = input("Enter distance traveled (in km): ")
    distance_km = float(distance_input)
    if distance_km < 0:
        print("Distance cannot be negative. Please try again.\n")
        continue

    peak_hour_input = input("Is it a peak hour? (yes/no): ").lower()
    is_peak_hour = False
    if peak_hour_input == 'yes':
        is_peak_hour = True
    elif peak_hour_input != 'no':
        print("Invalid input for peak hour. Please enter 'yes' or 'no'.\n")
        continue

    fare = calculate_cab_fare(distance_km, is_peak_hour)
    print(f"\nEstimated Cab Fare: ₹{fare:.2f}")
    break
```

## 5. Employee Attendance Eligibility Checker

```python
def check_attendance_eligibility(attendance_list):
    if not attendance_list:
        return "No attendence recorded."

    present_days = attendance_list.count('P')
    total_days = len(attendance_list)
    attendance_percentage = (present_days / total_days) * 100

    print(f"Attendence: {attendance_list}")
    print(f"Attendance Percentage: {attendance_percentage:.2f}%")

    if attendance_percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
while True:
    attendance_input = input("Enter attendence record (e.g., P P A P P, or type 'quit' to exit): ")
    if attendance_input.lower() == 'quit':
        break
    attendance_list = [day.strip().upper() for day in attendance_input.split() if day.strip()]
    if not all(day in ['P', 'A'] for day in attendance_list):
        print("Invalid input. Please use 'P' for Present and 'A' for Absent, separated by spaces.\n")
        continue

    print("--- Employee Status ---")
    result = check_attendance_eligibility(attendance_list)
    print(result)
    print("\n")
```

## 6. Password Strength Checker


```python
def check_password_strength(password):
    if len(password) < 8:
        return False, "Password must be at least 8 characters long."
    has_digit = False
    for char in password:
        if char.isdigit():
            has_digit = True
            break
    if not has_digit:
        return False, "Password must contain at least one digit."
    has_special_char = False
    special_chars = "@#$"
    for char in password:
        if char in special_chars:
            has_special_char = True
            break
    if not has_special_char:
        return False, f"Password must contain at least one special character from {special_chars}."

    return True, "Password strength: Strong."

while True:
    user_password = input("Enter your password (or type 'quit' to exit): ")
    if user_password.lower() == 'quit':
        break

    is_strong, message = check_password_strength(user_password)

    if is_strong:
        print(message)
    else:
        print(f"Password strength: Weak - {message}")
    print("\n")
```

## 1.  Employee Performance Bonus Eligibility


```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
```

```python
highest_score = 0
for score in employees.values():
    if score > highest_score:
        highest_score = score

top_performers = []
for employee, score in employees.items():
    if score == highest_score:
        top_performers.append(employee)
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
```

## 2. Search Query Keyword Analysis







```python
import collections
import string

search_query = "Buy mobile phone buy phone online"
processed_query = search_query.lower()
translator = str.maketrans('', '', string.punctuation)
processed_query = processed_query.translate(translator)
keywords = processed_query.split()
keyword_counts = collections.Counter(keywords)
result = {keyword: count for keyword, count in keyword_counts.items() if count > 1}
print(result)
```

## 3. Sensor Data Validation




```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for i, reading in enumerate(sensor_readings):
    if reading % 2 == 0:
        valid_readings.append((i, reading))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## 4. Email Domain Usage Analysis


```python
from collections import Counter
emails = [ "ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com" ]
domain_counts = Counter()
for email in emails:
    domain = email.split('@')[1]
    domain_counts[domain] += 1
total_emails = len(emails)
print("Email Domain Usage:")
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

## 5. Sales Spike Detection


```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
spike_threshold = average_sales * 1.30
print("Sales Spikes Detected:")
for i, sale in enumerate(sales):
    if sale > spike_threshold:
        print(f"Day {i + 1}: {sale}")
```

## 6. Duplicate User ID Detection


```python
import collections
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
user_id_counts = collections.Counter(user_ids)
print("Duplicate User IDs:")
for user_id, count in user_id_counts.items():
    if count > 1:
        print(f"{user_id} -> {count} times")
```

## 1. User Login Check

```python
username = "admin"
password = "1234"

input_username = input("Enter youur Username: ")
input_password = input("Enter your Password: ")

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

## 2. Pass / Fail Analyzer



```python
marks = [45, 78, 90, 33, 60]
pass_cnt = 0
fail_cnt = 0
for mark in marks:
   if mark >= 50:
    pass_cnt += 1
   else:
    fail_cnt += 1

print("Pass Count: ", pass_cnt)
print("Fail Count: ", fail_cnt)
```

## 3. Simple Data Cleaner



```python
names = [" Alice ", "bob", " CHARLIE "]
clean_naames = [name.strip().lower() for name in names]
print(clean_naames)
```

## 4. Message Length Analyzer


```python
messages = ["Hi", "Welcome to the platform", "OK"]
for msg in messages:
    length = len(msg)
    if length > 10:
      print(f"mesage: {msg} of length: {length}. (flag :message is longer than 10)!! ")
    else:
      print(f"mesage: {msg} of length: {length}. OK")
```

## 5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
err_cnt = 0
for log in logs:
    if log == "ERROR":
        err_cnt += 1

print("Error Count: ", err_cnt)
```

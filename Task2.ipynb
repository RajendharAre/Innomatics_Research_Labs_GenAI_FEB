## 1. Unique Words in a Sentence


```python
sentence = input("Enter a sentence: ")
words = sentence.split()
unique_words = []
for word in words:
    if word not in unique_words:
        unique_words.append(word)

print("Unique words in the sentence: ", unique_words)
```

## 2. Highest Salary from Employee Data







```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_salary = 0
highest_salary_employee = ""
for employee, salary in employees.items():
    if salary > highest_salary:
      highest_salary = salary
      highest_salary_employee = employee
print("Highest Salary:", highest_salary_employee, "-", highest_salary)
```

## 4. Count Products Above a Price Threshold



```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000: ", count)
```

## 5. Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]
present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100
print(f"Attendance Percentage: {percentage:.1f}")
```

## 6. Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_phone_numbers = set(phone_numbers)
print("Unique phone numbers:", unique_phone_numbers)
```

## 7. Count Character Frequency

```python
input_string = input("Enter a word: ")
char_freq = {}
for char in input_string:
    char_freq[char] = char_freq.get(char, 0) + 1
print(char_freq)
```

## 9. Check if a Key Exists in a Dictionary

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
search_key = "Anita"
if search_key in employees:
    print(f"{search_key} exists")
else:
    print(f"{search_key} does not exist")
```

## 10. Calculate Average Marks

```python
marks = [80, 65, 75, 70, 45, 34.5]
total_marks = sum(marks)
no_of_subjects = len(marks)
if no_of_subjects > 0:
    average_marks = total_marks / no_of_subjects
    print(f"Average Marks: {average_marks:.1f}")
else:
    print("No marks to calculate average.")
```
